```python
import numpy as np
from PIL import Image
from copy import deepcopy
import torch
import torch.optim as optim
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import DataLoader, TensorDataset, random_split
from torchvision.transforms.v2 import Compose, ToImage, Normalize, ToPILImage, Resize, ToDtype
from torchvision.datasets import ImageFolder
from torch.optim.lr_scheduler import StepLR, ReduceLROnPlateau, MultiStepLR, CyclicLR, LambdaLR
import matplotlib.pyplot as plt
device = 'cuda' if torch.cuda.is_available() else 'cpu'
```

```python
from torchvision import datasets, transforms
from torch.utils.data import DataLoader, WeightedRandomSampler
import numpy as np
```

```python
train_transform = transforms.Compose([
    transforms.RandomHorizontalFlip(),
    transforms.RandomRotation(15),
    transforms.RandomResizedCrop(64, scale=(0.7, 1.0)),
    transforms.ColorJitter(brightness=0.3, contrast=0.3, saturation=0.3),
    transforms.ToTensor(),
    transforms.Normalize((0.5,), (0.5,))
])

test_transform = transforms.Compose([
    transforms.Resize((64, 64)),
    transforms.ToTensor(),
    transforms.Normalize((0.5,), (0.5,))
])

train_data = datasets.ImageFolder(root='train', transform=train_transform)
val_data = datasets.ImageFolder(root='test', transform=test_transform)

labels = train_data.targets  # این یه لیست از کلاس‌ها (0 یا 1)

# تعداد نمونه هر کلاس
class_sample_count = np.array([np.sum(np.array(labels) == t) for t in np.unique(labels)])

# وزن هر کلاس برعکس تعداد نمونه‌ها
weight = 1. / class_sample_count

# وزن هر نمونه برابر با وزن کلاس مربوطه است
samples_weight = np.array([weight[t] for t in labels])

samples_weight = torch.from_numpy(samples_weight).double()
sampler = WeightedRandomSampler(samples_weight, len(samples_weight))
#, sampler=sampler
train_loader = DataLoader(train_data, batch_size=100, sampler=sampler)
valid_loader = DataLoader(val_data, batch_size=100, shuffle=False)
```

```python
class CNN2(nn.Module):
    def __init__(self, n_feature, p=0.0):
        super(CNN2, self).__init__()
        self.n_feature = n_feature
        self.p = p
        # Creates the convolution layers
        self.conv1 = nn.Conv2d(in_channels=3, out_channels=n_feature, kernel_size=3, padding=1)
        self.bn1 = nn.BatchNorm2d(n_feature)
        self.conv2 = nn.Conv2d(in_channels=n_feature, out_channels=n_feature*2, kernel_size=3, padding=1)
        self.bn2 = nn.BatchNorm2d(n_feature*2)
        self.fc1 = nn.Linear(n_feature*2 * 16 * 16, 200) 
        self.bn4 = nn.BatchNorm1d(200)
        self.fc2 = nn.Linear(200, 2)
        self.drop = nn.Dropout(self.p)
    
    def featurizer(self, x):
        x = self.conv1(x)
        x = self.bn1(x)
        x = F.relu(x)
        x = F.max_pool2d(x, kernel_size=2)
        x = self.conv2(x)
        x = self.bn2(x)
        x = F.relu(x) 
        x = F.max_pool2d(x, kernel_size=2)

        x = torch.flatten(x, start_dim=1) 
        return x
    
    def classifier(self, x):
        x = self.fc1(x)
        x = self.bn4(x)
        x = F.relu(x)
        if self.p > 0:
            x = self.drop(x)
        x = self.fc2(x)
        return x
    
    def forward(self, x):
        x = self.featurizer(x)
        x = self.classifier(x)
        return x
```

```python
torch.manual_seed(13)
model = CNN2(n_feature=32, p=0.5).to(device)
loss_fn = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(),  lr=0.001)
```

```python
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
```

```python
def accuracy(outputs, targets):
    correct = 0
    _, predicted = torch.max(outputs, 1)
    correct = (predicted == targets).sum().item()
    return correct / targets.size(0)

def train(model, dataloader, optimizer,criterion):
    #th = 0.45
    model.train()
    total_loss = 0.0
    total_acc = 0.0
    sum_precision=0.0
    sum_recall = 0.0
    sum_fscore = 0.0
    all_labels = []
    all_preds = []
    for inputs, targets in dataloader:
        optimizer.zero_grad()
        inputs, targets = inputs.to(device), targets.to(device)
        outputs = model(inputs)
        loss = criterion(outputs, targets) 
        _, predicted = torch.max(outputs, 1)
        all_labels.extend(targets.cpu().numpy())
        all_preds.extend(predicted.cpu().numpy())
        precision= precision_score(all_labels, all_preds, average="binary",zero_division=0.0)
        recall= recall_score(all_labels, all_preds, average="binary",zero_division=0.0)
        fscore=f1_score(all_labels, all_preds, average="binary",zero_division=0.0)
        loss.backward()
        optimizer.step()
        total_loss += loss.item()
        total_acc += accuracy(outputs, targets)
        sum_precision += precision
        sum_recall += recall
        sum_fscore += fscore
        #print("Predicted Positives train:", (outputs > 0.5).sum().item())
        #print("Predicted Negatives train:", (outputs <= 0.5).sum().item())
    return total_loss / len(dataloader),total_acc / len(dataloader),sum_precision / len(dataloader),sum_recall / len(dataloader),sum_fscore / len(dataloader),outputs

def validate(model, dataloader, criterion):
    #th = 0.45
    model.eval()
    total_loss = 0.0
    total_acc = 0.0
    sum_precision=0.0
    sum_recall = 0.0
    sum_fscore = 0.0
    all_labels = []
    all_preds = []
    with torch.no_grad():
        for inputs, targets in dataloader:
            inputs, targets = inputs.to(device), targets.to(device)
            outputs = model(inputs)
            #l1_regularization = 0
            l2_regularization = 0
            for param in model.parameters():
                 l2_regularization += torch.norm(param,2)
            loss = criterion(outputs, targets) + 0.0001*l2_regularization 
            _, predicted = torch.max(outputs, 1)
            all_labels.extend(targets.cpu().numpy())
            all_preds.extend(predicted.cpu().numpy())
            precision= precision_score(all_labels, all_preds, average="binary",zero_division=0.0)
            recall= recall_score(all_labels, all_preds, average="binary",zero_division=0.0)
            fscore=f1_score(all_labels, all_preds, average="binary",zero_division=0.0)
            total_loss += loss.item()
            total_acc += accuracy(outputs, targets)
            sum_precision += precision
            sum_recall += recall
            sum_fscore += fscore
    return total_loss / len(dataloader),total_acc / len(dataloader),sum_precision / len(dataloader),sum_recall / len(dataloader),sum_fscore / len(dataloader),outputs

```

```python
n_epochs = 10
train_losses , test_losses = [],[]
train_accuracies, test_accuracies = [], []
train_precisions, test_precisions = [], []
train_recalls, test_recalls = [], []
train_fscores, test_fscores = [], []

for epoch in range(n_epochs):
    train_loss, train_acc, train_perc, train_recall, train_fscore,output_train = train(model,train_loader,optimizer,loss_fn)
    train_losses.append(train_loss)
    train_accuracies.append(train_acc)
    train_precisions.append(train_perc)
    train_recalls.append(train_recall)
    train_fscores.append(train_fscore)
    test_loss, test_acc, test_perc, test_recall, test_fscore,output_test = validate(model,valid_loader,loss_fn)
    test_losses.append(test_loss)
    test_accuracies.append(test_acc)
    test_precisions.append(test_perc)
    test_recalls.append(test_recall)
    test_fscores.append(test_fscore)
    print(f"Epoch {epoch+1}/{n_epochs}, Train Loss: {train_loss:.4f}, Validation Loss: {test_loss:.4f}, "
          f"Train Accuracy: {train_acc:.4f}, Validation Accuracy: {test_acc:.4f}, \n"
          f"Train percision: {train_perc:.4f}, Validation percision: {test_perc:.4f},"
          f"Train recall: {train_recall:.4f}, Validation recall: {test_recall:.4f}, \n"
          f"Train fscore: {train_fscore:.4f}, Validation fscore: {test_fscore:.4f}, \n")

plt.figure(figsize=(12, 5))
plt.subplot(3, 2, 1)
plt.plot(range(1, n_epochs + 1), train_losses, label='Train Loss')
plt.plot(range(1, n_epochs + 1), test_losses, label='Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Training and Validation Losses')
plt.legend()

plt.subplot(3, 2, 2)
plt.plot(range(1, n_epochs + 1), train_accuracies, label='Train Accuracy')
plt.plot(range(1, n_epochs + 1), test_accuracies, label='Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.title('Training and Validation Accuracies')
plt.gca().set_yticklabels(['{:.0f}%'.format(x*100) for x in  plt.gca().get_yticks()])
plt.legend()

plt.subplot(3, 2, 3)
plt.plot(range(1, n_epochs + 1), train_precisions , label='Train Percision')
plt.plot(range(1, n_epochs + 1), test_precisions, label='Validation Percision')
plt.xlabel('Epoch')
plt.ylabel('Percision')
plt.title('Training and Validation Percisions')
plt.gca().set_yticklabels(['{:.0f}%'.format(x*100) for x in  plt.gca().get_yticks()])
plt.legend()

plt.subplot(3, 2, 4)
plt.plot(range(1, n_epochs + 1), train_recalls, label='Train Recall')
plt.plot(range(1, n_epochs + 1), test_recalls, label='Validation Recall')
plt.xlabel('Epoch')
plt.ylabel('Recall')
plt.title('Training and Validation Recalls')
plt.gca().set_yticklabels(['{:.0f}%'.format(x*100) for x in  plt.gca().get_yticks()])
plt.legend()

plt.subplot(3, 2, 5)
plt.plot(range(1, n_epochs + 1), train_fscores, label='Train fscore')
plt.plot(range(1, n_epochs + 1), test_fscores, label='Validation fscore')
plt.xlabel('Epoch')
plt.ylabel('fscore')
plt.title('Training and Validation fscore')
plt.gca().set_yticklabels(['{:.0f}%'.format(x*100) for x in  plt.gca().get_yticks()])
plt.legend()

plt.tight_layout()
plt.show()
```
